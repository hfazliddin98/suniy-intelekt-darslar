```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("https://github.com/anvarnarz/praktikum_datasets/raw/main/automobile_data_nan.csv", index_col=0)
df.head()
```

```python
df.fillna(0) # bunda NaN qiymatlar orniga 0 qiymatini koyib beradi
```

```python
df.fillna({'horsepower':100, 'wheel-base':90}) #bunda malum bir ustunlari bosh joylariga malum bir qiymat qoyiladi
```

```python
df.fillna(method='ffill') # bunda NaN qiymatlarni ozidan oldingi qiymatlar bilan toldirib beradi
```

```python
df.fillna(method='bfill') # bunda NaN qiymatini ozidan keyin keladigan qiymat bilan toldiradi
```
