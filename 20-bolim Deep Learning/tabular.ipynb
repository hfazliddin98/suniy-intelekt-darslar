#Tabular Data

```python
import pandas
from fastai.tabular.all import *
```

```python
# datasets
path = untar_data(URLs.ADULT_SAMPLE)

path.ls()
```

```python
# dataloders
dls = TabularDataLoaders.from_csv(path/'adult.csv', path=path, bs=64, y_names='salary',
                                  # data ichidagi bosh joylarni olib tashlaydi
                                  skipinitialspace=False,
                                  cat_names=['workclass', 'education', 'marital-status', 'occupation', 'relationship', 'race', 'sex'],
                                  cont_names=['age', 'fnlwgt', 'education-num', 'capital-gain', 'capital-loss', 'hours-per-week'],
                                  # bunda data boyicha etaplarni yani tozalashlarni togridan togri yozishimiz mumkun
                                  procs=[Categorify, FillMissing, Normalize])
```

```python
# o`qitish (train)
learn = tabular_learner(dls, metrics=accuracy)
# fastai ozi suniy neronlar qilib beradi bu yerda biz nechta epokdan qilishini korsatishimiz kerak
learn.fit_one_cycle(5)
```

```python
# datasetni chaqirib olamiz
import pandas as pd

df = pd.read_csv(path/'adult.csv')
df.head()
```

```python
# test qilamiz
row, pred, probs = learn.predict(df.iloc[0])
pred.item(), probs
```
