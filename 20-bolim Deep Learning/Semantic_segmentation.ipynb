### datasetni yuklab olish

```python
from fastai.vision.all import *

path = untar_data(URLs.CAMVID_TINY)
# fallar ro`yhati
path.ls()
```

### label getter fuksiya yaratib oliamiz

```python
def label_getter(file):
  return path/'labels'/f'{file.stem}_P{file.suffix}'

# dataloader
dls = SegmentationDataLoaders.from_label_func(path, bs=8, fnames=get_image_files(path/'images'), label_func=label_getter,
                                              codes=np.loadtxt(path/'codes.txt', dtype=str))
# fine_tune
learn = unet_learner(dls, resnet34)
learn.fine_tune(12)
```

```python
# rasmlarni chiqarib korishimiz munkun oldingi va yaratgan rasmlarni
learn.show_results(max_n=5, figsize=(16,15))
```
