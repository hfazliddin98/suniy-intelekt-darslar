```python
x = np.arange(0, 4*np.pi, 0.1)
y = np.sin(x)
```

```python
plt.plot(x,y)
plt.show()
```

```python
plt.plot(x, y, linewidth=3) # chiziq qalinligi hisoblanadi
plt.show()
```

```python
plt.plot(x, y, linewidth=3, linestyle='--') # bunda chiziq qalinliga va chiziq qanday holatida bolishini kiritishimiz ninkun
plt.show()
```

```python
plt.plot(x, y, linewidth=3, linestyle='--', color='#8D87AD') # bunda chiziq qalinligi ,stili, va rangini belgilashimiz munkun qanday bolishini
plt.show()
```

```python
plt.plot(x, y, linestyle='--', marker='o') # bunda chiziqqa malum bir shakl joylashtirishmiz munkun
plt.show()
```

```python
plt.figure(figsize=(10,4)) # girafik boyi va enini o`lchamini berishimiz munkun 

plt.plot(x, y, marker='o')
plt.title('sin(x) fuksiya') # grafikka nom berishimiz munkun
plt.xlabel('pi') # x o`qini qiymatiga nom beramiz
plt.ylabel('sin(x)') # y o`qiga qiymatiga nom beramiz
plt.grid() # bunda x va y oqlaridagi chiziqlarni birlashtirib  yani setka chizib beradi
plt.show()
```

```python
x = np.arange(0, 4*np.pi, 0.1)
sin = np.sin(x)
cos = np.cos(x)

# grafig o`lchami
plt.figure(figsize=(12,6))

# sin(x) grafigi
plt.plot(x, sin, linewidth=4, linestyle=':', alpha=0.7, marker='h')

# cos(x) grafigi
plt.plot(x, cos, linewidth=4, linestyle=':', alpha=0.7, marker='v')

# grafig nomi 
plt.title('sin(x) va  cos(x) funksiya grafigi ')

# x o`qi nomi
plt.xlabel('pi')

# har bir chiziq haqida malumot
plt.legend(['sin(x)', 'cos(x)'], loc=0, fontsize='x-large')

# setkani o`zgartirish
plt.grid(color='tomato', alpha=0.5, linestyle='--') 

# ramini saqlab olish uchun ishlatiladi
plt.savefig('rasmlar/sin.png')
plt.show()
```
