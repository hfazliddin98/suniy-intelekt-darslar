```python
import pandas as pd
import numpy as np
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns # agar o`rnatilmagan bolsa pip install seaborn
```

```python
# parse_dates datefildga o`zgartiradi
df = pd.read_csv('https://github.com/anvarnarz/praktikum_datasets/raw/main/museum_visitors.csv', index_col=0, parse_dates=[0], usecols=['Date', 'Avila Adobe'])
df.head()
```

```python
df['month'] = [i.month for i in df.index]
df.head()
```

```python
df['year'] = [i.year for i in df.index]
df.head()
```

```python
df1 = df.groupby(['month', 'year']).max()
df1.head()
```

```python
df2 = df1.unstack(level=0) # indexlarini bitasini ustun ikinchisini qatorga aylantirb beradi
```

```python
plt.figure(figsize=(12,6))
sns.heatmap(data=df2, cmap='YlGnBu', linewidths=1, square=True)
plt.show()
```
