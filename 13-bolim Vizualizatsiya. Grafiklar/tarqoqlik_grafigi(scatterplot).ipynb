```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

## Scatter plot

```python
df = pd.read_csv('https://raw.githubusercontent.com/anvarnarz/praktikum_datasets/main/merc.csv')
df.head()
```

```python
plt.figure(figsize=(10,6))
sns.scatterplot(data=df, x='year', y='price') # year va price o`rtasidagi bogliqlik yani koreliyatsiyani korishimiz munkun
plt.show()
```

```python
df = df[df['year']>2000] # bunda 2000-yildan tepada bolgan mashinalarni olib qolindi
plt.figure(figsize=(10,6))
sns.scatterplot(data=df, x='year', y='price') # year va price o`rtasidagi bogliqlik yani koreliyatsiyani korishimiz munkun
plt.show()
```
