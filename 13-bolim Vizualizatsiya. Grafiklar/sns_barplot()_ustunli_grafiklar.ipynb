```python
import pandas as pd
import numpy as np
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns # agar o`rnatilmagan bolsa pip install seaborn
```

```python
df = pd.read_csv('https://raw.githubusercontent.com/anvarnarz/praktikum_datasets/main/uzbekistan.csv', index_col=0)
df.columns = ['maydon', 'aholi']
df.head()
# df.shape
```

```python
plt.figure(figsize=(16,5))
sns.barplot(x=df.index, y=df.aholi)
plt.xlabel('Hudud')
plt.title('O`zbekiston Respublikasi tarkibidagi hududlarning aholi soni')
plt.ylabel('Aholi soni (million)')
plt.grid()
plt.xticks(rotation=90)
plt.show()
```

```python
plt.figure(figsize=(16,5))
sns.barplot(x=df.index, y=df.maydon)
plt.xlabel('Hudud')
plt.title('O`zbekiston Respublikasi tarkibidagi hududlarning aholi soni')
plt.ylabel('Maydoni kv.km ')
plt.grid()
plt.xticks(rotation=90)
plt.show()
```
