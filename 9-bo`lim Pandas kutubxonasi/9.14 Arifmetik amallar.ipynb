```python
from numpy.core.fromnumeric import reshape
df1 = pd.DataFrame(np.arange(9.).reshape(3,3), columns=list('bcd'), index=['Olma','Anor','Uzum'])
df2 = pd.DataFrame(np.arange(12.).reshape(4,3), columns=list('abc'), index=['Olma','Anor','Qovun','Anjir'])
```

```python
df1
```

```python
df2
```

```python
df1+df2
```

```python
df1.add(df2) # dataframni bir biriga qoshish
```

```python
df1.add(df2, fill_value=0) # ikkisidan birida bor bolsa 0 qoshib chiqaradi 
```

```python
df1.sub(df2, fill_value=0) # Dataframlarni bir biridan ayirish uchun ishlatiladi 
```

```python
df1.rsub(df2, fill_value=0) # dataframlarni o`rnini almashtirib ayiradi
```

```python
df1.div(df2) # dataframlarni bir birig bolish uchun ishlatiladi
```
