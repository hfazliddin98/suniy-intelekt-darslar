```python
import pandas as pd
import pprint as print
import sqlite3 as sql
```

tartiblash(ORDER BY) va saralsh(DISTINCT)

```python
# Database ga bog`lanish uchun`
db = sql.connect('fayllar/chinook.db')
cursor = db.cursor()
```

```python
# Malumotlarni ajratib olish uchun
komanda = "SELECT CustomerId, FirstName, LastName, Country,Phone,Email FROM customers;"
df = pd.read_sql(komanda, db, index_col='CustomerId')
df.head(20) 
```

```python
# malumotlarni tartiblash uchun
komanda = "SELECT CustomerId, FirstName, LastName, Country,Phone,Email FROM customers ORDER BY FirstName;"
df = pd.read_sql(komanda, db, index_col='CustomerId')
df.head()
```

```python
# bazadagi malum bir ustuni malumotini takrorlamasdan saralab olish
komanda = "SELECT DISTINCT Country FROM customers;"
cursor.execute(komanda)
cursor.fetchall()
```

```python
# DataFrame korinishida olish uchun
df = pd.read_sql(komanda,db)
df
```
