```python
A = pd.DataFrame({'Trasport':['poyezd', 'avtomobil', 'kema'], 
                  'Tezligi':['70','180','45']})
A
```

```python
B = pd.DataFrame({'Trasport':['poyezd', 'avtomobil', 'tranvay'], 
                  'Narxi':['98000','10000','65000']})
B
```

```python
birlash= pd.merge(A, B) # bunda ikki dataframni birlashtiradi
birlash
```

```python
marget = pd.merge(A, B, how='inner') # inner join icki - bunda ikkita dataframda ham bor umumiy malumotlarni birlashtiriladi
marget
```

```python
merget_outer = pd.merge(A, B, how='outer') #outer join tashqi bunda ikkita dataframni hamma malumotlari birlashtiriladi
merget_outer
```

```python
merget_left = pd.merge(A, B, how='left') # left join chap bunda chap datafremga birlashtiradi va chap dataframda yo`qlarini tashlab yuboradi
merget_left
```

```python
merget_right = pd.merge(A, B, how='right') # right join o`ng bunda o`ng datafremga birlashtiradi va o`ng dataframda yo`qlarini tashlab yuboradi
merget_right
```
